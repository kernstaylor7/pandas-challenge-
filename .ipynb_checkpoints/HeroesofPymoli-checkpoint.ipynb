```python
import pandas as pd
import numpy as np
purchase_data= pd.read_csv("purchase_data.csv")

purchase_data.head()

                  
```

```python

total_players = purchase_data['SN'].nunique()
print(total_players)


```

```python

items = purchase_data['Item Name'].nunique()
avg_price= round(purchase_data['Price'].mean(), 2)
total_purchases= purchase_data['Purchase ID'].count()
total_revenue=purchase_data['Price'].sum()


analysis= {'Purchases': [total_purchases],
           'Average Price': [avg_price],
           'Total Revenue': [total_revenue],
           'Unique Items': [items]}

analysis = pd.DataFrame (analysis, columns = ['Purchases', 'Average Price', 'Total Revenue', 'Unique Items'])

print(analysis)
                
```

```python
df=purchase_data.drop_duplicates('SN') 
male_players= df.loc[df['Gender']=='Male'].count()['Gender']
male_percent= round((male_players / total_players).sum() * 100, 2)

female_players=df.loc[df['Gender']=='Female'].count()['Gender']
female_percent= round((female_players / total_players).sum() * 100, 2)

other_players=df.loc[df['Gender']=='Other / Non-Disclosed'].count()['Gender']
other_percent= round((other_players / total_players).sum() * 100, 2)


gender_demographics = {'Total Count': [male_players, female_players, other_players],
        'Percent': [male_percent, female_percent, other_percent]}
  
gender_demographics = pd.DataFrame (gender_demographics,[['Male', 'Female', 'Other/Non-Disclosed']], columns = ['Total Count','Percent'])

print (gender_demographics)
```

```python
pc_male = purchase_data.loc[purchase_data['Gender'] == 'Male', :].count()['Price']
avg_pc_male = round(purchase_data.loc[purchase_data['Gender'] == 'Male', :].mean()['Price'],2)
total_pc_male = round(purchase_data.loc[purchase_data['Gender'] == 'Male', :].sum()['Price'],2)
avg_tp_male = round(total_pc_male/male_players,2)

pc_female = purchase_data.loc[purchase_data['Gender'] == 'Female', :].count()['Price']
avg_pc_female = round(purchase_data.loc[purchase_data['Gender'] == 'Female', :].mean()['Price'],2)
total_pc_female = round(purchase_data.loc[purchase_data['Gender'] == 'Female', :].sum()['Price'],2)
avg_tp_female = round(total_pc_female/female_players,2)


pc_other = purchase_data.loc[purchase_data['Gender'] == 'Other / Non-Disclosed', :].count()['Price']
avg_pc_other = round(purchase_data.loc[purchase_data['Gender'] == 'Other / Non-Disclosed', :].mean()['Price'],2)
total_pc_other = round(purchase_data.loc[purchase_data['Gender'] == 'Other / Non-Disclosed', :].sum()['Price'],2)
avg_tp_other = round(total_pc_other/other_players,2)

purchasing_ana = {'Purchase Count': [pc_male, pc_female, pc_other],
        'Avg Purchase Price': [avg_pc_male, avg_pc_female, avg_pc_other],
        'Total Purchase Value': [total_pc_male, total_pc_female, total_pc_other],
        'Avg Purchase Totals per Person':[avg_tp_male, avg_tp_female, avg_tp_other]}

purchasing_ana = pd.DataFrame (purchasing_ana,[['Male', 'Female', 'Other/Non-Disclosed']], columns = ['Purchase Count','Avg Purchase Price', 'Total Purchase Value', 'Avg Purchase Totals per Person' ])
print(purchasing_ana)
```

```python
bins = [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 100]

age_ranges = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']

age_groups = df.loc[:,['SN','Age']]
age_groups.drop_duplicates(['SN'], inplace=True, keep='first')

age_groups['Age Group'] = pd.cut(age_groups['Age'], bins, labels = age_ranges)

summary = pd.DataFrame()
summary['Total Count'] = age_groups.groupby(['Age Group'])['Age'].count()
summary['% of Players']=(age_groups.groupby(['Age Group'])['Age'].sum()/576)*100
summary['% of Players']=summary['% of Players'].map("%{:,.2f}".format)
summary
```

```python
top_spenders=df.groupby("SN")
spender_count=round(top_spenders["Purchase ID"].count(),2)
avg_spender=round(top_spenders["Price"].mean(),2)
total_spend= round(top_spenders["Price"].sum(),2)

spender_summary = pd.DataFrame({"Purchase Count": top_spenders,
                                    "Avg Purchase Price": avg_spender,
                                    "Total Purchase Value": total_spend})
spender_summary = spender_summary.sort_values(["Total Purchase Value"], ascending=False).head(5)

spender_summary["Avg Purchase Price"] = spender_summary["Avg Purchase Price"].map("${:.2f}".format)
spender_summary["Total Purchase Value"] = spender_summary["Total Purchase Value"].map("${:.2f}".format)

```
